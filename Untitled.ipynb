# Problem #1 :- Creating a boolean mask to convert all the numbers greater than 50 to zeros in the array.

```python
numbers = np.array(range(1, 101))
print(f'Array with numbers from 1 to 100: {numbers}')


# Create a boolean mask to convert the number greater than 50 to zero
mask_50 = numbers > 50
numbers[mask_50] = 0
print(f'Numbers greater than 50 as zeros: {numbers}')
print(f'Correct answer #1')

mask_50_2 = np.where(numbers > 50)
numbers[mask_50_2] = 0
print(f'Numbers greater than 50 as zeros: {numbers}')
print(f'Correct answer #2')
```

```python
#Creating an array that stores the marks of the students
scores = np.array([35, 55, 38, 92, 45, 28, 88])

#Creating a condition for pass and fail and toppers
conditions = [
    scores < 40,
    (scores >= 40) & (scores < 90),
    scores >= 90
]

#Creating the variables that will replace the correct values
pft = ["fail", "pass", "topper"]

#Finally adding them all as a function
result = np.select(conditions, pft, default="Unknown")
print(f'Final array with pass, fail and toppers: {result}')
```
